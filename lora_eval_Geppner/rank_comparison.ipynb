```python
# Re-import necessary libraries after execution state reset
import pandas as pd
import matplotlib.pyplot as plt
prompt_dict={0:"<phigep>",1:"<phigep> on a parkbench, wearing pink crocs",2:"<phigep> playing chess in space"}
# Load the first dataset (Rank 16)
df_rank_16 = pd.read_csv("res_florencer16.csv")
df_rank_16=df_rank_16.groupby(["loraname", "prompt_id", "training_step"]).mean().reset_index().sort_values(["prompt_id","training_step"])
df_rank_16=df_rank_16[df_rank_16["training_step"]<=2400]
# Apply rolling average with window size 3
df_rank_16_rolling = df_rank_16.copy()
for col in ["eval_same_person_llm", "eval_limb_quality_llm", "eval_artifacts_quality_llm", "eval_prompt_adherence_llm","eval_image_quality_llm"]:
    df_rank_16_rolling[col] = df_rank_16_rolling[col].rolling(window=3, min_periods=1).mean()

# Create a 1x3 grid for Same Person, Limb Quality, Artifacts, and Prompt Adherence comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

for i, prompt_id in enumerate(df_rank_16["prompt_id"].unique()):
    ax = axes[i]
    # Filter data for the given prompt_id
    df_prompt = df_rank_16_rolling[df_rank_16_rolling["prompt_id"] == prompt_id]
    # Plot each metric
    ax.plot(df_prompt["training_step"], df_prompt["eval_same_person_llm"],
             color='mediumseagreen', marker='o', linestyle='-', label="Same Person")
    ax.plot(df_prompt["training_step"], df_prompt["eval_limb_quality_llm"],
             color='darkviolet', marker='s', linestyle='--', label="Limb Quality")
    ax.plot(df_prompt["training_step"], df_prompt["eval_image_quality_llm"],
             color='deepskyblue', marker='^', linestyle='-.', label="Image Quality")
    ax.plot(df_prompt["training_step"], df_prompt["eval_prompt_adherence_llm"],
             color='tomato', marker='d', linestyle='-', label="Prompt Adherence")

    ax.set_xlabel("Training Step")
    ax.set_title(f"Prompt {prompt_dict[prompt_id]}")
    ax.set_xticks(range(200, df_rank_16["training_step"].max() + 200, 200))
    ax.set_ylim(0, 1)  # Set y-axis limits

axes[0].set_ylabel("Evaluation Score")
fig.suptitle("Comparison of Same Person, Limb Quality, Image Quality  & Prompt Adherence - Rolling Avg (Window=3)", fontsize=14)

# Add a single legend at the bottom
fig.legend(["Same Person", "Limb Quality", "Image Quality", "Prompt Adherence"], loc="lower center", ncol=4, bbox_to_anchor=(0.5, -0.15))

plt.grid(True)
plt.show()
```
